<b>3:</b> 3.1 describes a de-synchronization attack, where an eavesdropper catches data values of A,B,C, alters them, sends it to the tag, and the tag sends out data values for D, E and the eavesdropper also changes those data values before they get to the tag, meaning it makes it more easier to decipher the ID in each passing exchange. This is a security concern because the tag sending out A,B,C is doing it in unencrypted manner, and the reader can't verify this when it gets the altered D, and E from the tag. Its only slightly different from the full-disclosure attack, because it starts out with deciphering n2, then using bits of the de-synchronization attack in later stages of the full-disclosure attack to help solve for ID in the E data value.

<div class="alert alert-block alert-success">
<b>4:</b> MMAP and EMAP use the IDP identifier to determine what K values are used.  Those K values and the IDP are used in every A-E equation.  The question of How can we remove dependency on the IDP is kind of like asking How can I remove all the wiring in my car and still have it run?  Sure you can do that, but your car wont work any more. In terms of reducing dependence on IDP, or reducing the vulnerability, some possible avenues would be to introduce a time-varying identification code similar to other RFID protocols or the EMV chip in a credit card, and setting a maximum number of protocol runs in which this number would be valid. This would, naturally, require changes to the MMAP/EMAP protocols, but would necesstiate fewer changes than eliminating dependence on the IDP entirely.
